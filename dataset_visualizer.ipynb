```python
import json
```

```python
def data_extract(d: dict):
    print(f"Title: {d['title']}")
    for pair in d["paragraphs"]:
        print("Context")
        print(pair["context"])
        print("")
        print("QAs")
        for qa in pair["qas"]:
            print(f"Q. {qa['question']}")
            print(f"A (idx: {qa['answers'][0]['answer_start']}). {qa['answers'][0]['text']}")
        print("")
```

```python
data_filename = "KorQuAD_v1.0_dev.json"
with open(data_filename, "r") as f:
    data = json.load(f)
    print(f"Dataset version: {data['version']}")
    print("")
    data_extract(data["data"][1])
```
